```python
data.drop_duplicates(keep='first',inplace=True)
```

```python
X=data.drop(columns="Result")
Y=data["Result"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
report = []
threshold = 0.05

for column in x_train.columns:
    result = ks_2samp(x_train[column], x_test[column])
    report.append({"column": column, "pvalue": result.pvalue, "drift_status": result.pvalue < threshold})

report_df = pd.DataFrame(report)
report_df
```

```python
imputer=KNNImputer()
x_train=imputer.fit_transform(x_train)
x_test=imputer.transform(x_test)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import f1_score, precision_score, roc_auc_score

models = {
    "Random Forest": RandomForestClassifier(random_state=42, verbose=1),
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "Gradient Boosting": GradientBoostingClassifier(random_state=42, verbose=1),
    "Logistic Regression": LogisticRegression(verbose=1),
    "AdaBoost": AdaBoostClassifier(random_state=42),
}

params = {
    "Decision Tree": {
        "criterion": ["gini", "entropy", "log_loss"],
    },
    "Random Forest": {
        "n_estimators": [8, 16, 128, 256]
    },
    "Gradient Boosting": {
        "learning_rate": [0.1, 0.01, 0.05, 0.001],
        "subsample": [0.6, 0.7, 0.75, 0.85, 0.9],
        "n_estimators": [8, 16, 32, 64, 128, 256]
    },
    "Logistic Regression": {},
    "AdaBoost": {
        "learning_rate": [0.1, 0.01, 0.001],
        "n_estimators": [8, 16, 32, 64, 128, 256]
    }
}

```

```python
report = []

for model_name, model in models.items():
    print(f"ðŸ” Training {model_name}...")
    
    param_grid = params.get(model_name, {})  
    gs = GridSearchCV(model, param_grid, cv=3, scoring="f1", verbose=1)  
    gs.fit(x_train, y_train)

    best_model = gs.best_estimator_
    best_model.fit(x_train, y_train)

    y_train_pred = best_model.predict(x_train)
    y_test_pred = best_model.predict(x_test)

    train_f1 = f1_score(y_train, y_train_pred, average="weighted")
    test_f1 = f1_score(y_test, y_test_pred, average="weighted")

    train_precision = precision_score(y_train, y_train_pred, average="weighted")
    test_precision = precision_score(y_test, y_test_pred, average="weighted")

    train_roc = roc_auc_score(y_train, y_train_pred, average="weighted", multi_class="ovr")
    test_roc = roc_auc_score(y_test, y_test_pred, average="weighted", multi_class="ovr")

    report.append({
        "model": model_name,
        "train_f1": train_f1,
        "test_f1": test_f1,
        "train_precision": train_precision,
        "test_precision": test_precision,
        "train_roc": train_roc,
        "test_roc": test_roc
    })

results = pd.DataFrame(report)
```
